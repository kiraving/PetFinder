```python
def xgb_progressbar(rounds=1000):
    """Progressbar for xgboost using tqdm library.
 
    Examples
    --------
 
    >>> model = xgb.train(params, X_train, 1000, callbacks=[xgb_progress(1000), ])
    """
    pbar = tqdm(total=rounds)
 
    def callback(_, ):
        pbar.update(1)
 
    return callback
```

```python
import cv2
import pandas as pd
import numpy as np
import os
from tqdm import tqdm, tqdm_notebook
from keras.applications.densenet import preprocess_input, DenseNet121, DenseNet169

#train_df = pd.read_csv('../input/petfinder-adoption-prediction/train/train.csv')
img_size = 256
batch_size = 16
pet_ids = train_df['PetID'].values
n_batches = len(pet_ids) // batch_size + 1

def resize_to_square(im):
    old_size = im.shape[:2] # old_size is in (height, width) format
    ratio = float(img_size)/max(old_size)
    new_size = tuple([int(x*ratio) for x in old_size])
    # new_size should be in (width, height) format
    im = cv2.resize(im, (new_size[1], new_size[0]))
    delta_w = img_size - new_size[1]
    delta_h = img_size - new_size[0]
    top, bottom = delta_h//2, delta_h-(delta_h//2)
    left, right = delta_w//2, delta_w-(delta_w//2)
    color = [0, 0, 0]
    new_im = cv2.copyMakeBorder(im, top, bottom, left, right, cv2.BORDER_CONSTANT,value=color)
    return new_im

def load_image(path, pet_id):
    image = cv2.imread(f'{path}{pet_id}-1.jpg')
    new_image = resize_to_square(image)
    new_image = preprocess_input(new_image)
    return new_image
```

```python
from keras.models import Model
from keras.layers import GlobalAveragePooling2D, Input, Lambda, AveragePooling1D
import keras.backend as K
inp = Input((256,256,3))
backbone = DenseNet121(
    #weights = r'./input/keras-applications-weights/densenet169_weights_tf_dim_ordering_tf_kernels_notop.h5',
    input_tensor = inp, include_top = False)
print("done")
x = backbone.output
x = GlobalAveragePooling2D()(x)
x = Lambda(lambda x: K.expand_dims(x,axis = -1))(x)
x = AveragePooling1D(4)(x)
out = Lambda(lambda x: x[:,:,0])(x)

m = Model(inp,out)
```

```python
features = {}
for b in tqdm(range(n_batches)):
    start = b*batch_size
    end = (b+1)*batch_size
    batch_pets = pet_ids[start:end]
    batch_images = np.zeros((len(batch_pets),img_size,img_size,3))
    #print(b)
    for i,pet_id in enumerate(batch_pets):
        try:
            batch_images[i] = load_image(r'./input/petfinder-adoption-prediction/train_images/', pet_id)
        except:
            pass
    batch_preds = m.predict(batch_images)
    for i,pet_id in enumerate(batch_pets):
        features[pet_id] = batch_preds[i]
```

```python
train_feats = pd.DataFrame.from_dict(features, orient='index')
train_feats.to_csv('train_img_features.csv')
train_feats.head()
```

```python

n_batches = len(test_df['PetID'].values) // batch_size + 1
features = {}
for b in tqdm(range(n_batches)):
    start = b*batch_size
    end = (b+1)*batch_size
    batch_pets = pet_ids[start:end]
    batch_images = np.zeros((len(batch_pets),img_size,img_size,3))
    #print(b)
    for i,pet_id in enumerate(batch_pets):
        try:
            batch_images[i] = load_image("../input/petfinder-adoption-prediction/test_images/", pet_id)
        except:
            pass
    batch_preds = m.predict(batch_images)
    for i,pet_id in enumerate(batch_pets):
        features[pet_id] = batch_preds[i]
test_feats = pd.DataFrame.from_dict(features, orient='index')
test_feats.to_csv('test_img_features.csv')
test_feats.head()
```

```python
tab_imfeatures = pd.concat([train_df, train_feats.reset_index(drop=True)], axis=1,sort=False)
train_df = tab_imfeatures

tab_imfeatures2 = pd.concat([test_df, test_feats.reset_index(drop=True)], axis=1,sort=False)
test_df = tab_imfeatures2
```

```python
train_df.head()
```

```python
permuted = train_df.reindex(np.random.permutation(train_df.index))
ytrain = permuted['AdoptionSpeed']
clean_train = permuted.drop(columns=['Name', 'RescuerID', 'Description', 'PetID', 'AdoptionSpeed'])
clean_test = test_df.drop(columns=['Name', 'RescuerID', 'Description', 'PetID'])

df_columns = clean_train.columns
num_train = len(clean_train)

x_train = clean_train[:num_train//10*9]
y_train = ytrain[:num_train//10*9]
x_val = clean_train[num_train//10*9:]
y_val = ytrain[num_train//10*9:]
print(df_columns)
dtrain = xgb.DMatrix(x_train.values, y_train.values)#, feature_names=df_columns)
dval = xgb.DMatrix(x_val.values)#, feature_names=df_columns)
dtest = xgb.DMatrix(clean_test.values)#, feature_names=df_columns)
```

```python
#настраиваем XGBoost
xgb_params = {'subsample': 0.6, 'n_estimators': 300, 'min_child_weight': 5, 
              'max_depth': 7, 'learning_rate': 0.03, 'gamma': 2, 'num_class':5,
              'colsample_bytree': 1.0, 'objective':'multi:softmax'}
def kappa(y1,y2): 
    score = cohen_kappa_score(y1,y2, weights='quadratic')
    return 'kappa', score


num_boost_round=500
```

```python
num_boost_round=500
model = xgb.train( dict(xgb_params),#subsample=0.9, n_estimators=500, min_child_weight=2, 
                           #max_depth=7, learning_rate=0.03, gamma=2, colsample_bytree=0.8),
                  dtrain,feval=kappa, maximize=True,
                   num_boost_round=num_boost_round,verbose_eval=False, callbacks=[xgb_progressbar(num_boost_round)])
```

```python
val_pred=model.predict(dval).astype(int)
metric(val_pred, y_val.values)
```

```python

test_pred=np.round(model.predict(dtest)).astype(int)
print(test_pred.max())
submit_df = pd.DataFrame(data={"PetID" : test_df["PetID"], 
                                   "AdoptionSpeed" : test_pred})
submit_df.to_csv("submission.csv", index=False)
submit_df.head()
```
